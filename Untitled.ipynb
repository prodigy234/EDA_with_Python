```python
import matplotlib.pyplot as plt

# Data for the chart
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']
values = [10, 15, 7, 12]

# Create the figure and axis
fig, ax = plt.subplots()

# Plot the bar chart
ax.bar(labels, values)

# Set title and labels
ax.set_title('Fruit Quantities')
ax.set_xlabel('Fruit')
ax.set_ylabel('Quantity')

# Show the plot
plt.show()
```
